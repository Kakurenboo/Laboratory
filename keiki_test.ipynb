```python
#ランダムフォレスト
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import warnings
from sklearn.utils.testing import all_estimators
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
import warnings
warnings.filterwarnings('ignore')
from dtreeviz.trees import dtreeviz


# 景気データの読み込み
keiki_data = pd.read_csv("keiki.csv", encoding="utf8")
target_names = np.array(["不景気", "好景気"], dtype="<U10")
keiki_target_name= "DII"

# 景気データをラベルと入力データに分離する 
y = keiki_data.loc[:,"DII"]
x = keiki_data.loc[:,["C1","C2","C3","C4","C5","C6","C7","C8","C9"]]

# 学習用とテスト用に分離する 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, train_size = 0.8, shuffle = True)

#データ変換
#lab_enc = preprocessing.LabelEncoder()
#y_trained = lab_enc.fit_transform(y_train)
#y_tested = lab_enc.fit_transform(y_test)

#木を作って学習させて評価
clf = RandomForestClassifier(n_estimators=100,random_state=0,max_depth=9)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
print("RandomForestClassifierの正解率 = " , accuracy_score(y_test, y_pred)*100,"%")

#視覚化
x_list=["C1","C2","C3","C4","C5","C6","C7","C8","C9"]
print(type(x_list))
#clf.fit(iris.data, iris.target)

if x_list is None:
    print('xはNoneTypeです')
else:
    print('大丈夫')


estimators = clf.estimators_
viz = dtreeviz(
    estimators[0],
    x_train,
    y_train,
    target_name=keiki_target_name,
    feature_names = x_list,
    class_names=[str(i) for i in target_names],
)

viz
```
